```python
import numpy as np
import pandas as pd

exam_data = {'name' : ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                      'Matthew', 'Laura', 'Kevin', 'Jonas'],

    'score' : [12.5, 9, np.NaN, 16.5 , 9, np.NaN, 20, 14.5, 8, 19],

    'attempts' : [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

    'qualify' : ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}


labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

exam_dataframe = pd.DataFrame(exam_data, index = labels)

print(exam_dataframe)
```

```python
#number 2. Delete rows with Nan values.

exam_dataframe.drop(exam_dataframe.loc[exam_dataframe['score'] == 'NaN'].index, inplace=True)

exam_dataframe
```

```python
#number 3. Extract the 'name' and 'score' columns from the DataFrame.

x = exam_dataframe.iloc[:,0:2]
y = exam_dataframe.iloc[:,[0,1]]
z = exam_dataframe[['name', 'score']]


x
```

```python
#number 4. Write a Pandas program to append a new row 'k' to the DataFrame with these values
#(name: "Suresh", score: 15.5, attempts: 1, qualify: "yes")

dict = {'name': "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}

label = ['k']

df = pd.DataFrame(dict, index = label)

exam_dataframe.append(df)

```

```python
#number 5. Write a Pandas program to delete the 'attempts' column from the DataFrame.

atte = exam_dataframe.drop('attempts', axis = 1)

atte
```

```python
#number 6. Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.

exam_dataframe['Success'] = [1 if x > 10 else 0 for x in exam_dataframe['score']]


print(exam_dataframe)
```
